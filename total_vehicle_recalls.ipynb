```python
# CSV_File_Reading
vrbd_60days_daily = pd.read_csv("/Users/prsm/Desktop/Fintech/Projects/VR group project/vrdb_60days_daily.csv")
vrbd_60days_daily.head(10)
```

```python
vrbd_60days_daily = vrbd_60days_daily.drop(columns=[ "CATEGORY_FTXT", "YEAR", "MANUFACTURER_RECALL_NO_TXT", "CATEGORY_ETXT", "UNIT_AFFECTED_NBR", "SYSTEM_TYPE_ETXT", "SYSTEM_TYPE_FTXT", "NOTIFICATION_TYPE_ETXT", "NOTIFICATION_TYPE_FTXT", "COMMENT_ETXT"])
```

```python
vrbd_60days_daily = vrbd_60days_daily[['MAKE_NAME_NM','MODEL_NAME_NM','RECALL_NUMBER_NUM']]
```

```python
vehicle_df = vrbd_60days_daily.groupby(['MAKE_NAME_NM']).count()
vehicle_df.head(10)
```

```python
vehicle_df = vehicle_df.drop(columns =["MODEL_NAME_NM"])
```

```python
vehicle_df
```

```python
vehicle_df = vehicle_df[vehicle_df.RECALL_NUMBER_NUM>25]
```

```python
vehicle_df.plot(kind = 'pie',x='MAKE_NAME_NM', y='RECALL_NUMBER_NUM', figsize = (15,15), title = 'Distribution of vehicle recall by make').legend(loc='upper right')
```
