```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Data collection from the repository
csv_path = Path (r"C:\Users\nnroc\OneDrive\Documents\Fintech Course\Project 1\Vehicle Recalls\Resources\vrdb_60days_daily.csv")
vehicle_recalls_df = pd.read_csv(csv_path)
vehicle_recalls_df.head()
```

```python
# data cleaning
# remove unwanted columns
vehicle_recalls_df = vehicle_recalls_df.drop(columns=["RECALL_NUMBER_NUM", "MANUFACTURER_RECALL_NO_TXT", "CATEGORY_FTXT", "SYSTEM_TYPE_FTXT", "NOTIFICATION_TYPE_FTXT", "COMMENT_ETXT", "COMMENT_FTXT"])
# remove duplicates
vehicle_recalls_df = vehicle_recalls_df.drop_duplicates().copy()
# remove , in the number fields
vehicle_recalls_df["UNIT_AFFECTED_NBR"] = vehicle_recalls_df["UNIT_AFFECTED_NBR"].str.replace(",", "")
# convert the data type of number of unit from string to float
vehicle_recalls_df["UNIT_AFFECTED_NBR"] = vehicle_recalls_df["UNIT_AFFECTED_NBR"].astype('float')
# convert the data type of number of unit from float to integer
vehicle_recalls_df["UNIT_AFFECTED_NBR"] = vehicle_recalls_df["UNIT_AFFECTED_NBR"].astype('int')
#display the sample data
vehicle_recalls_df.head()
```

```python
# specify the column that contains the recall date
date_col = 'RECALL_DATE_DTE'

# sort the dataframe by the recall date in descending order
vehicle_recalls_df = vehicle_recalls_df.sort_values(by=date_col, ascending=False)

# display the most recent recalls
display(vehicle_recalls_df.head())
```

```python
# select only recall date and make name
vehicle_recalls_df = vehicle_recalls_df[["RECALL_DATE_DTE", "MAKE_NAME_NM"]]

# sorted data by year
vehicle_recalls_df = vehicle_recalls_df.sort_values("RECALL_DATE_DTE", ascending = False)

#display the sample data
vehicle_recalls_df.head()
```

```python
vehicle_recalls_df = vehicle_recalls_df.groupby(["RECALL_DATE_DTE"]).count()
```

```python
vehicle_recalls_df = vehicle_recalls_df.sort_values("RECALL_DATE_DTE", ascending = False)
```

```python
vehicle_recalls_df.head() 
```

```python
vehicle_recalls_df = vehicle_recalls_df.reset_index()
```

```python
vehicle_recalls_df = vehicle_recalls_df[vehicle_recalls_df.RECALL_DATE_DTE> '2022-09-30']
```
